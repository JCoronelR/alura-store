### Generando Graficos

#### Grafico ingreso por tienda

```python
import matplotlib.pyplot as plt

# Datos de ingreso por tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [115088040.0, 111634350.0, 109801960.0, 103837570.0]

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(tiendas, ingresos, color='blue', edgecolor='black')

# Añadir etiquetas de valor encima de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1000000, f'{yval:,.0f}', ha='center', va='bottom', fontsize=10)

# Títulos y etiquetas
plt.title('Ingresos Totales por Tienda', fontsize=14)
plt.xlabel('Tiendas')
plt.ylabel('Ingreso Total ')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

## **Informe Final**

El propósito de este análisis es asesorar al Sr. Juan sobre cuál de las tiendas evaluadas es la mejor opción para comercializar sus productos, considerando variables como ingresos totales, popularidad de categorías, satisfacción de clientes, productos con mayor y menor demanda y coste de envío promedio. El objetivo es maximizar sus beneficios y minimizar riesgos operativos, seleccionando la opción que brinde mayor estabilidad, mejores márgenes y proyección de ventas positiva.

#### Ingresos Totales por Tienda
De acuerdo con los gráficos y análisis generados, se pudo observar que la Tienda 1 presentó los ingresos totales más altos durante el periodo analizado, superando ampliamente a Tienda 2,  Tienda 3 y Tienda 4. Esta variable es crucial ya que mayores ingresos reflejan mayor volumen de ventas y, potencialmente, una mayor cartera de clientes.

#### Categorías de Productos Más y Menos Vendidas
Los datos muestran que, en la Tienda 1, 2, 3, 4 la categoría de "Muebles" fue la más vendida, mientras que "Articulos del Hogar" e "Instrumentos musicales" tuvo menor demanda. Estas tendencias reflejan distintos perfiles de clientela y pueden marcar oportunidades o riesgos dependiendo del catálogo que el Sr. Juan maneje.

#### Calificaciones Promedio de los Clientes por Tienda
Las calificaciones promedio por tienda muestran que la Tienda 3 obtuvo la valoración más alta de los clientes, seguida por la Tienda 2 y Tienda 4, en último lugar la Tienda 1. Un mayor promedio de calificaciones indica buena satisfacción del cliente y menos riesgo de conflictos post-venta.

#### Productos Más y Menos Vendidos
En cuanto a productos específicos, el análisis identificó que el producto más vendido en la Tienda 1 corresponde a un bien de alta demanda recurrente, mientras que en la Tienda 4 los productos más vendidos están sujetos a la estacionalidad, reflejando mayor volatilidad.

#### Coste de Envío Promedio por Tienda
Finalmente, el coste de envío promedio por tienda fue más favorable en la Tienda 4, mientras que la Tienda 1, Tienda 2 y Tienda 3 presentaron valores similares y ligeramente superiores. Un coste de envío bajo suele incentivar mayores compras, mejorando recurrencia y satisfacción del cliente.

#### Conclusión y Recomendación
Tras considerar todos los factores evaluados y respaldar cada argumento con las visualizaciones y análisis de datos, la opción óptima para el Sr. Juan es la Tienda 1. Esta tienda destaca claramente por sus altos ingresos totales y volumen de ventas, lo que garantiza una mayor visibilidad y acceso a un mercado amplio. Aunque la Tienda 4 sobresale en  coste de envío, su volumen de ingresos es considerablemente menor, lo que podría limitar las posibilidades de crecimiento a largo plazo. Tienda 4 presenta factores de riesgo por su estacionalidad y menores calificaciones de clientes.

Por tanto, la recomendación fundamentada es que el Sr. Juan venda la tienda 4, ya que ella es la que tiene menos fortalezas que las demas tiendas.
